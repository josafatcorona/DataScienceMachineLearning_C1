```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f= pd.read_csv('data/bank.csv')
f.head()
```

```python
f[['duration','y']]
```

```python
#We drop column duration because if we need to predict wether a client is going to buy a product, we can not know that field
f.drop('duration',axis=1,inplace=True)
f.shape
```

```python
#Split x and y
X = f.iloc[:,:-1]
X.head()
```

```python
y = f.iloc[:,-1]
y.head()
```

```python
#Create variable dummies
X = pd.get_dummies(X, drop_first=True)
y = pd.get_dummies(y,drop_first=True)
X.head()
```

```python
y.head()
```

```python
#Split the dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1234)
print(f"X_train:{X_train.shape} X_test:{X_test.shape} y_train:{y_train.shape} y_test:{y_test.shape}")
```

```python
#Import random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#Default Random Foresr Object
rfc1 = RandomForestClassifier(random_state=1234)
rfc1.fit(X_train,y_train)
y_predict1 = rfc1.predict(X_test)
```

```python
#Score and evaluate the model
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test,y_predict1)
cm1
```

```python
score1 = rfc1.score(X_test,y_test)
score1
```

## Now we are going to implement Recursive Feature Elimination

```python
#Apply Recursive Featrue Elimination and select 30 Featrues
from sklearn.feature_selection import RFE
```

```python
#estimator is an algorithm
#Get and object Estimator
rfc2 = RandomForestClassifier(random_state=1234)
rfe = RFE(estimator=rfc2,n_features_to_select=30)
rfe.fit(X,y)
```

```python
x_train_rfe= rfe.transform(X_train)
x_test_rfe = rfe.transform(X_test)
print(f"X_test:{x_test_rfe.shape} X_train:{x_train_rfe.shape}")
```

```python
x_train_rfe[5]
```

```python
#Default Random Foresr Object
rfc2.fit(x_train_rfe,y_train)
y_predict2 = rfc2.predict(x_test_rfe)
```

```python
#Score and evaluate the model
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(y_test,y_predict2)
cm2
```

```python
score2 = rfc2.score(x_test_rfe,y_test)
score2
```

```python
#We have removed 22 features and got the same almost result
#get columns names
columns = list(X.columns)
```

```python
#get ranking feature importance
ranking = rfe.ranking_
ranking
#all 1 are the selected columns
```

```python
# get the feature importance
feature_importance = rfc1.feature_importances_
feature_importance
```

```python
# Create the dataframe of the Features selected, Ranking and their importance
rfe_selected = pd.DataFrame()
rfe_selected = pd.concat([pd.DataFrame(columns), 
                          pd.DataFrame(ranking),
                          pd.DataFrame(feature_importance)], axis=1)

rfe_selected.columns = ["Feature Name", "Ranking", "Feature Importance"]
```
